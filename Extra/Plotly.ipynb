```python
df = pd.read_excel('4-GoogleLatLng/GoogleLatLng.xlsx')
```

```python
df = df[['bairro', 'categoria', 'preco']]
df.groupby(['categoria'])['preco'].count()
```

```python
df_casa = df[df['categoria'] == 'Casas']
df_casa = df_casa[['bairro', 'preco']]
df_casa = df_casa[df_casa['preco'] > 100000]
df_casa.reset_index(drop=True, inplace=True)
```

```python
df_casa
```

```python
#cria um data frame com a soma dos precos dos municipios
df_casa_soma = df_casa.groupby(['bairro'], as_index=False)['preco'].sum()
df_casa_soma.columns= ['bairro', 'preco_soma']
#media
df_casa_media = df_casa.groupby(['bairro'], as_index=False)['preco'].mean()
df_casa_media.columns= ['bairro', 'preco_media']
#numero de ocorrencias
df_casa_count = df_casa.groupby(['bairro'],as_index=False)['preco'].count()
df_casa_count.columns= ['bairro', 'count']
```

```python
df_agrupa = df_casa_soma.merge(df_casa_media, on='bairro')
df_agrupa = df_agrupa.merge(df_casa_count, on='bairro')
```

```python
df_agrupa = df_agrupa.sort_values('preco_media', ascending=False).reset_index(drop=True)
df_agrupa.head()
```

```python
data = [
    go.Bar(
        x=df_agrupa['bairro'].head(),
        y=df_agrupa['preco_media'].head()
    )
]
```

```python
# IPython notebook
py.plotly.iplot(data, filename='OLX_SP_CASAS_MEDIA_x_BAIRRO', title='SP-CASAS À VENDA SOMA x BAIRRO')
```

```python
df_agrupa = df_agrupa.sort_values('preco_soma', ascending=False).reset_index(drop=True)
data = [
    go.Bar(
        x=df_agrupa['bairro'].head(),
        y=df_agrupa['preco_soma'].head()
    )
]
```

```python
# IPython notebook
py.plotly.iplot(data, filename='OLX_SP_CASAS_SOMA_x_BAIRRO', title='SP-CASAS À VENDA SOMA x BAIRRO')
```
